```python
#1. Tạo một DataFrame từ dữ liệu
import pandas as pd
import pandas as np
import matplotlib.pyplot as plt


data = {
    'Name': ['Mr. Harris', 'Mrs. Bradley', 'Miss. Laina', 'Mrs. Heath', 'Mr. William', 'Mr. James',
             'Mr. Timothy', 'Mrs. Oscar', 'Mrs. Nicholas', 'Miss. Marguerite', 'Mr. William',
             'Mrs. Hulda', 'Mr. Lawrence', 'Mrs. Wilma'],
    'Height': [1.80, 1.61, 1.78, 1.74, 1.55, 1.9, 1.61, 1.40, 1.92, 1.77, 1.55, 1.85, 1.64, 1.77],
    'Weight': [90, 87, 64, 70, 100, 106, 82, 55, 77, 60, 100, 109, 80, 66],
    'Age': [32, 45, 22, 52, 48, 21, 28, 69, 58, 21, 48, 33, 25, 70]
}

df = pd.DataFrame(data)
```

```python
#2. Hiển thị thông tin về DataFrame vừa tạo
print(df)
```

```python
#3. Tính giá trị trung bình của cột <AGE>.
average_age = df['Age'].mean()

print(f"Gia tri trung binh cot Age: {average_age}")
```

```python
#4. Nhóm theo cột 'Height' và tính trung bình cột 'Weight'
grouped = df.groupby('Height')['Weight'].mean().reset_index()

print(grouped)
```

```python
#5. Sắp xếp theo cột 'Weight' tăng dần
df_sorted = df.sort_values(by='Weight', ascending=True)

print(df_sorted)
```

```python
#6. Vẽ biểu đồ cột cho cột 'Age'
plt.figure(figsize=(10,6))
plt.bar(df['Name'], df['Age'], color='skyblue')
plt.xlabel('Name')  #x
plt.ylabel('Age')  #y
plt.title('Bieu do cot Ages')
plt.xticks(rotation=90)
plt.show()
```

```python
#7. Vẽ biểu đồ phân tán cho cột 'Age' và 'Height'
plt.figure(figsize=(10,6))
plt.scatter(df['Age'], df['Height'], color='orange')
plt.xlabel('Age')
plt.ylabel('Height')
plt.title('Phan tan cot Age va Height')
plt.show()
```

```python
#8. Vẽ biểu đồ đường cho cột 'Weight'
plt.figure(figsize=(10,6))
plt.plot(df['Name'], df['Weight'], marker='o', color='b', linestyle='-', markersize=6)  # Biểu đồ đường
plt.xlabel('Name')
plt.ylabel('Weight')
plt.title('Bieu do duong cot Weight')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

```python
# 9. Vẽ biểu đồ tròn cho cột 'Height'
plt.figure(figsize=(6,6))
plt.pie(df['Height'], labels=df['Name'], autopct='%1.1f%%', startangle=140)
plt.title('Bieu do tron cot Height')
plt.show()
```

```python
# 10. Vẽ biểu đồ phân tán cho cột 'Weight' và 'Age'
plt.figure(figsize=(8,5))
plt.scatter(df['Age'], df['Weight'], color='b')
plt.xlabel('Age')
plt.ylabel('Weight')
plt.title('Bieu do phan tan cot Age va Weight')
plt.show()
```

```python
#11. Kiểm tra có giá trị NaN trong DataFrame không
has_nan = df.isna().any().any()

# In kết quả
if has_nan:
    print("DataFrame có giá trị NaN.")
else:
    print("DataFrame không có giá trị NaN.")
```

```python
#12. Tìm các giá trị trùng lặp trong DataFrame dựa trên cột 'Name' và loại bỏ các hàng trùng lặp, giữ lại hàng đầu tiên.
df_unique = df.drop_duplicates(subset='Name', keep='first')

print(df_unique)
```

```python
#13. Thay thế các giá trị của cột <Age> lớn hơn 40 bằng giá trị trung bình của cột đó
average_age = int(df['Age'].mean())

df.loc[df['Age'] > 40, 'Age'] = average_age

print(df)
```

```python
#14. Tính chỉ số BMI và thêm vào cột 'cs_BMI'
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)

print(df)
```

```python
#15. Tạo cột với tên <Phan_Loai> với các mức ‘0’,’1’,’2’,’3’ tương ứng

df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)

# Hàm phân loại BMI
def phan_loai(bmi):
    if bmi < 18.5:
        return 0  # Thiếu cân
    elif 18.5 <= bmi <= 22.9:
        return 1  # Bình cân
    elif 23 <= bmi <= 24.9:
        return 2  # Thừa cân
    else:
        return 3  # Béo phì

# Áp dụng hàm phân loại vào cột 'cs_BMI' và tạo cột 'Phan_Loai'
df['Phan_Loai'] = df['cs_BMI'].apply(phan_loai)

# Hiển thị kết quả
print(df)
```

```python
# 16. Chuẩn hóa giá trị của cột <cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1.
# Tính chỉ số BMI và thêm vào cột 'cs_BMI'
df['cs_BMI'] = df['Weight'] / (df['Height'] ** 2)

# Chuẩn hóa giá trị của cột 'cs_BMI' trong đoạn từ 0 tới 1
min_bmi = df['cs_BMI'].min()  # Giá trị BMI nhỏ nhất
max_bmi = df['cs_BMI'].max()  # Giá trị BMI lớn nhất

df['cs_BMI_normalized'] = (df['cs_BMI'] - min_bmi) / (max_bmi - min_bmi)

# Hiển thị kết quả
print(df)
```

```python
#17. Tạo cột với tên <GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong cột <Name> còn các từ đó.
# Tạo cột GT dựa trên giá trị trong cột Name
df['GT'] = df['Name'].apply(lambda x: 'male' if 'Mr' in x else ('female' if 'Mrs' in x or 'Miss' in x else 'unknown'))

# Xóa các từ 'Mr', 'Mrs', 'Miss' khỏi cột Name
df['Name'] = df['Name'].str.replace('Mr. ', '').str.replace('Mrs. ', '').str.replace('Miss. ', '')

# Hiển thị kết quả
print(df)
```

```python
#18. Vẽ biểu đồ tròn cho cột 'GT'
gt_counts = df['GT'].value_counts()  # Tính số lượng mỗi giới tính

# Vẽ biểu đồ tròn
plt.figure(figsize=(6, 6))
plt.pie(gt_counts, labels=gt_counts.index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcoral'])
plt.title('Giới Tính (GT) Phân Bố')
plt.show()
```

```python
#19. Trực quan hóa dữ liệu người nữ trên <Phan_Loai>.
df_female = df[df['GT'] == 'female']

phan_loai_counts = df_female['Phan_Loai'].value_counts()

plt.figure(figsize=(8, 6))
phan_loai_counts.plot(kind='bar', color='lightgreen', edgecolor='black')
plt.title('Phân Loại BMI cho Nhóm Nữ')
plt.xlabel('Phân Loại BMI')
plt.ylabel('Số Lượng')
plt.xticks(rotation=0)
plt.show()
```
